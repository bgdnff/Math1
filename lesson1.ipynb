```python
import math


a = 5
b = 7
# Окружность:
r = 5
x = []
y1 = []
y2 = []
for i in range(1001):
    x_=i/100-5
    x.append(x_)
    y1.append(math.sqrt(r**2-x_**2))
    y2.append(-math.sqrt(r**2-x_**2))
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel('x')
plt.ylabel('y')
    
```

```python
# ЭЛЛИПС
a = 5
b = 10
x = []
y1 = []
y2 = []
for i in range(1001):
    x_=i/100-5
    x.append(x_)
    y1.append(b*math.sqrt(1-(x_/a)**2))
    y2.append(-b*math.sqrt(1-(x_/a)**2))
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel('x')
plt.ylabel('y')
    
```

```python
# ГИПЕРБОЛА
a = 2
b = 1
x = []
y1 = []
y2 = []
for i in range(1001):
    x_=i/100-5
    x.append(x_)
    y1.append(b*math.sqrt(1+(x_/a)**2))
    y2.append(-b*math.sqrt(1+(x_/a)**2))
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel('x')
plt.ylabel('y')
```

```python
x = np.linspace(-2, 3, 201)
plt.plot(x, (np.exp(x) +x- 1)/x)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1, 8) 
plt.grid(True)
plt.show()

from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x*(1-y) - 1)

x1, y1 =  fsolve(equations, (-2, 1))         
print (x1, y1)

x2, y2 =  fsolve(equations, (2, 6))              
print (x2, y2)
```
